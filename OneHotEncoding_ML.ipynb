```python
import pandas as pd
```

```python
df = pd.read_csv('/homeprices.csv')
df
```

```python
print(df.isnull().sum())
```

Using pandas to create dummy variables

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['town'],axis='columns')
final
```

dummy variable trap

```python
final= final.drop(['monroe township'],axis='columns')
final
```

```python
X = final.drop('price',axis='columns')
```

```python
X
```

```python
y=final.price
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict(X)
```

```python
model.score(X,y)
```

```python
model.predict([[3400,0,0]])
```

```python
model.predict([[2800,0,1]])
```

Using sklearn on OneHotEncoder
label encoder to convert town names into numbers

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle=df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
X = dfle[['town','area']].values
```

```python
X
```

```python
y=dfle.price.values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
# Use 'remainder' instead of 'remainder'
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder='passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
```

```python
X
```

```python
model.fit(X,y)
```

```python
model.predict([(0,1,3300)])
```

```python
model.predict([[1,0,2200]])
```
